## 데이터 전처리

```python
want_data_seoul=data_seoul[(data_seoul['상권업종대분류코드']=='D')&(data_seoul['상권업종중분류명']=='종합소매점')]
df1=want_data_seoul[(want_data_seoul['상권업종소분류명']=='수퍼마켓')]
df2=want_data_seoul[want_data_seoul['상권업종소분류명']=='편의점']
df3=want_data_seoul[want_data_seoul['상권업종소분류명']=='종합소매']
df4=want_data_seoul[want_data_seoul['상권업종소분류명']=='할인점']
want_data_seoul=pd.concat([df1, df2, df3, df4], axis=0)

want_data_busan=data_busan[(data_busan['상권업종대분류코드']=='D')&(data_busan['상권업종중분류명']=='종합소매점')]
df1_1=want_data_busan[want_data_busan['상권업종소분류명']=='수퍼마켓']
df2_1=want_data_busan[want_data_busan['상권업종소분류명']=='편의점']
df3_1=want_data_busan[want_data_busan['상권업종소분류명']=='종합소매']
df4_1=want_data_busan[want_data_busan['상권업종소분류명']=='할인점']
want_data_busan=pd.concat([df1_1, df2_1, df3_1, df4_1], axis=0)
```

```python
select_seoul_store=want_data_seoul[['상가업소번호','상호명','상권업종소분류명','시도명','시군구명','행정동명','지번주소','도로명주소','층정보','경도','위도']]
select_busan_store=want_data_busan[['상가업소번호','상호명','상권업종소분류명','시도명','시군구명','행정동명','지번주소','도로명주소','층정보','경도','위도']]

select_seoul_store=select_seoul_store.reset_index()
select_busan_store=select_busan_store.reset_index()
```

## 데이터 불러오기
* 지역별 상권업종소분류별 인구 데이터

```python
seoul_data=pd.read_csv('./data/select_seoul.csv',encoding='cp949')
person_seoul = pd.read_csv("./data/자치구단위 서울 생활인구 일별 집계표.csv",encoding='euc-kr')
```

```python
df = seoul_data.groupby(['상권업종소분류명','시군구명']).count()[['상가업소번호']].copy()
df['ind'] = df.index
df['상권업종소분류명'] = df['ind'].apply(lambda x: x[0])
df['시군구명'] = df['ind'].apply(lambda x: x[1])
df = df.drop(columns=['ind']).copy()
df.index = range(len(df))
df.columns = ['상가_개수','상권업종소분류명','시군구명']
df = df[['상권업종소분류명','시군구명','상가_개수']].copy()
```

```python
gu_list = person_seoul['시군구명'].unique().tolist()
df['인구수'] = 0
for i in gu_list:
    df.loc[df['시군구명']==i,'인구수'] = person_seoul.loc[person_seoul['시군구명']==i,'총생활인구수'].iloc[0]
```

```python
work = df['상권업종소분류명'].unique().tolist()
df['비율'] = df['인구수']/df['상가_개수']
ratio_mean = df.groupby("상권업종소분류명").mean()[['비율']].copy()
ratio_mean['상권업종소분류명'] = ratio_mean.index
ratio_mean.index = range(len(ratio_mean))
```

```python
df['평균비율'] = 0
for i in work:
    df.loc[df['상권업종소분류명']==i,'평균비율'] = ratio_mean.loc[ratio_mean['상권업종소분류명']==i,'비율'].iloc[0]
```

```python
df['tf'] = df['비율']<df['평균비율']
check = df[df['tf']==True].copy()
check.groupby('시군구명').count().sort_values(['상가_개수'])[['상가_개수']]
check.groupby(["상권업종소분류명","시군구명"]).mean()[['비율','평균비율']]
```

```python
df.to_csv('people_seoul.csv',encoding='euc-kr')
```
